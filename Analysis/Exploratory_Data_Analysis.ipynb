```python
# Importing necessary libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Loading the dataset into a pandas DataFrame
df = pd.read_csv("Agrofood_co2_emission.csv")
df
```

```python
# Displaying a concise summary of the DataFrame
# This includes the number of entries, columns, data types, and memory usage
df.info()
```

```python
# Generating descriptive statistics for numeric columns in the DataFrame
# This includes count, mean, std, min, max, and quartile values
df.describe()
```

```python
# Checking for missing values in the DataFrame
# The isna() method identifies all missing (NaN) values in the DataFrame.
# The sum() method aggregates the count of missing values for each column.
df.isna().sum()
```

```python
# Replacing all missing (NaN) values in the DataFrame with 0
df.fillna(0,inplace = True)
```

```python
#Rename the total_emission column
df.rename(columns= {"total_emission": "total_emission_with_forestland"}, inplace = True)

#Calculate total emissions excluding the "Forestland" emissions.
df["total_emission_without_forestland"] = df[
    [
        "Savanna fires",
        "Forest fires",
        "Crop Residues",
        "Rice Cultivation",
        "Drained organic soils (CO2)",
        "Pesticides Manufacturing",
        "Food Transport",
        "Net Forest conversion",
        "Food Household Consumption",
        "On-farm Electricity Use",
        "Food Packaging",
        "Agrifood Systems Waste Disposal",
        "Food Processing",
        "Fertilizers Manufacturing",
        "IPPU",
        "Manure applied to Soils",
        "Manure left on Pasture",
        "Manure Management",
        "Fires in organic soils",
        "Fires in humid tropical forests",
        "On-farm energy use",
    ]
].sum(axis=1)
```

```python
# Defining a list of emission sources related to fires
fires = [ 'Savanna fires', 'Forest fires', 'Fires in organic soils', 'Fires in humid tropical forests']

# Defining a list of emission sources related to agricultural practices
agricultural_practices = [
    'Crop Residues', 'Rice Cultivation', 'Drained organic soils (CO2)', 
    'Manure applied to Soils', 'Manure left on Pasture', 'Manure Management']

# Defining a list of emission sources related to agri-food system practices
agri_food_system_related_practices = [
    'Pesticides Manufacturing', 'Food Transport', 'Food Household Consumption', 'Food Retail', 
    'On-farm Electricity Use', 'Food Packaging', 'Agrifood Systems Waste Disposal', 
    'Food Processing', 'Fertilizers Manufacturing', 'IPPU', 'On-farm energy use']

print("Fires:", fires)
print("\nAgricultural Practices:", agricultural_practices)
print("\nAgri-Food System-Related Practices:", agri_food_system_related_practices)
```

#### Trends in Average Total Emissions Over Time: With and Without Forestland

```python
# Grouping the DataFrame by the "Year" column
# This groups the data by each year in order to calculate the mean emission 
emission_data = df.groupby("Year")[["total_emission_without_forestland", "total_emission_with_forestland"]].mean()

# Styling
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Directly plot the entire DataFrame
emission_data.plot(ax=plt.gca(), linewidth=2)

# Title and labels
plt.title("Trends in Total Emissions (Yearly Averages)", fontsize=16, pad=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Emissions (Units)", fontsize=12)
plt.legend(title="Emission Categories", fontsize=10, loc="upper left", bbox_to_anchor=(1.05, 1))
plt.grid(visible=True, which="major", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Calculating the mean value of the "Forestland" column for each area
df.groupby("Area")[["Forestland"]].mean().sort_values(by="Forestland", ascending= True).head(10).reset_index()
```

```python
# Filter data for "Russian Federation" and calculate mean emissions by year
russia_emission_data = (
    df[df["Area"] == "Russian Federation"]
    .groupby("Year")[["total_emission_without_forestland", "total_emission_with_forestland"]]
    .mean()
)

# Styling
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Directly plot the filtered DataFrame
russia_emission_data.plot(ax=plt.gca(), linewidth=2)

# Title and labels
plt.title("Trends in Total Emissions for the Russian Federation (Yearly Averages)", fontsize=16, pad=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Emissions (Units)", fontsize=12)
plt.legend(title="Emission Categories", fontsize=10, loc="upper left", bbox_to_anchor=(1.05, 1))
plt.grid(visible=True, which="major", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```

####  How is the % of agri food co2 emission related to total emission?Â¶

```python
# Calculate the total emissions from the agrifood sector in 2019, convert the result to gigatons, 
# and determine its percentage of the estimated global emissions.

y_2019 = df.loc[df["Year"]==2019]
kt_2019 =y_2019["total_emission_with_forestland"].sum()
gt_2019 = round(kt_2019 / 1_000_000,2) # the amount of gigatones of CO2
perc = gt_2019 / 30

print(f"The amount of CO2 from agrifood in 2019 is {gt_2019} gigatones (gt), that is:",round(perc * 100), "%","of total emissions!")
```

#### How do CO2 emissions vary over time? 

```python
# Group data by year and calculate the sum of emissions
emissions_by_year = df.groupby("Year")[["total_emission_with_forestland"]].sum()

# Styling
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))

# Plot using seaborn's lineplot
sns.lineplot(data=emissions_by_year, x="Year", y="total_emission_with_forestland", linewidth=2)

# Title and labels
plt.title("Total CO2 Emissions Over Time (Yearly Totals)", fontsize=16, pad=20)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Emissions (Units)", fontsize=12)

# Grid customization
plt.grid(visible=True, which="major", linestyle="--", linewidth=0.5)

# Ensure layout fits well
plt.tight_layout()
plt.show()
```

#### Which regions contribute the most to total CO2 emissions?


```python
# Visualizing the top 20 regions with the highest total CO2 emissions, including forestland contributions.

# Group the data by region ("Area") and calculate the total emissions for each region
emissions_by_region = df.groupby("Area")[["total_emission_with_forestland"]].sum().sort_values(by = "total_emission_with_forestland", ascending = False)

# Extract the top 20 regions with the highest emissions and reset the index for visualization
top_20_emissions_by_region = emissions_by_region[:20].reset_index()

# Create a horizontal bar plot to visualize the top 20 regions
colors = plt.cm.get_cmap("coolwarm_r", len(top_20_emissions_by_region))
plt.figure(figsize= (10,6))
plt.barh(top_20_emissions_by_region["Area"],top_20_emissions_by_region["total_emission_with_forestland"], color = colors(range(len(top_20_emissions_by_region))))
plt.title("Top 20 Regions by Total C02 Emissions")
plt.xlabel("Total Emissions")
plt.ylabel("Region")
plt.gca().invert_yaxis()
plt.show()
```

#### How the CO2 emissions change per capita over time?

```python
# Filter required columns
data = df[['Area', 'Year', 'total_emission_without_forestland']]

# Calculate cumulative emissions over time for each area exculuding forestland contributions
data = data.sort_values(by=['Area', 'Year'])
data['cumulative_emission'] = data.groupby('Area')['total_emission_without_forestland'].cumsum()

# Create an animated map
fig = px.choropleth(
    data,
    locations="Area",
    locationmode="country names",
    color="cumulative_emission",
    hover_name="Area",
    animation_frame="Year",
    title="Cumulative COâ Emissions Over Time",
    color_continuous_scale=px.colors.sequential.Oranges,
    labels={"cumulative_emission": "Cumulative COâ Emissions (kt)"},
    range_color=(0, data['cumulative_emission'].max())
)

# Update layout for better visualization
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type="equirectangular",
        showcountries=True,
        countrycolor="white",  # Set country boundaries initially white
        showland=True,
        landcolor="lightgrey"  # Areas with no data will be grey
    ),
    coloraxis_colorbar=dict(
        title="Cumulative COâ Emissions",
        ticksuffix=" kt"
    ),
    margin=dict(l=0, r=0, t=50, b=0),
    width=1000,
    height=600
)

# Adjust color to fill over time and start unfilled
fig.update_traces(
    marker_line_width=0,
    colorscale=[[0, "white"], [1, px.colors.sequential.Oranges[-1]]],
    zmin=0,
    zmax=data['cumulative_emission'].max()
)

# Show the map
fig.show()
```

#### How does population growth correlate with C02 emissions?

```python
# Set figure size
plt.figure(figsize=(14, 8))

# Group data by year and calculate the sum for populations and emissions
pop_emission_data = df.groupby("Year")[["Rural population", "Urban population", "total_emission_with_forestland"]].sum()

# Plot total emissions with specific styling
ax = pop_emission_data["total_emission_with_forestland"].plot(
    label="Total CO2 Emissions", 
    secondary_y=False, 
    color="red", 
    linestyle="--", 
    linewidth=2
)

# Plot rural and urban population on the secondary y-axis
pop_emission_data[["Rural population", "Urban population"]].plot(ax=ax, secondary_y=True, linewidth=2)

# Add title, labels, and legend
plt.title("Population Growth and CO2 Emissions Over Time", fontsize=16, pad=20)
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("Total CO2 Emissions (Units)", fontsize=12)
ax.right_ax.set_ylabel("Population (Count)", fontsize=12)
ax.legend(["Total CO2 Emissions", "Rural Population", "Urban Population"], fontsize=10, loc="upper left", bbox_to_anchor=(1.05, 1))

# Customize grid
plt.grid(visible=True, which="major", linestyle="--", linewidth=0.5)

# Ensure layout fits neatly
plt.tight_layout()
plt.show()
```
